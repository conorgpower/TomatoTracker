```
import cv2
import tensorflow as tf

categories = ['Tomato_Septoria_leaf_spot', 
              'Potato___Early_blight', 
              'Tomato_Spider_mites_Two_spotted_spider_mite',
              'Pepper__bell___healthy']

def prepare(image_directory):
    image_size = 256
    image_array = cv2.imread(image_directory)
    image_array = cv2.resize(image_array, (256, 256))
    return image_array.reshape(-3, image_size, image_size, 3)

model = tf.keras.models.load_model("/content/drive/My Drive/my_model")
```

```
prediction = model.predict([prepare('/content/drive/My Drive/input/plantdisease/PlantDisease/Potato___Early_blight/001187a0-57ab-4329-baff-e7246a9edeb0___RS_Early.B 8178.JPG')])
print(categories[int(prediction[0][0])])
```
